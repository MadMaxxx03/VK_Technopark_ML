```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import sparse
from tqdm import tqdm
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
import hdbscan
```

**Из всех зачений в format только 1 не пустое, а в shape - 2**

```python
data = sparse.load_npz("train.npz")
```

**Пробую понять оптимальное колличество кластеров**

```python
n = 30
dist = []
silhouette_scores = []

for k in tqdm(range(2, n)):
    model = KMeans(n_clusters=k)
    model.fit(data)
    
    dist.append(model.inertia_)

    silhouette_scores.append(silhouette_score(data, model.labels_))
```

```python
plt.plot(range(2, n), dist)
plt.xlabel('Количество кластеров')
plt.ylabel('Дистанция')
plt.title('График метода локтя')
plt.grid(True)
plt.show()
```

```python
plt.plot(range(2, n), silhouette_scores)
plt.xlabel('Количество кластеров')
plt.ylabel('Silhouette score')
plt.title('График silhouette score')
plt.grid(True)
plt.show()
```

```python
def make_sub(labels):
    submission = pd.DataFrame({'ID':range(data.shape[0]), 'TARGET':labels})
    submission.to_csv('submission.csv', index=False)
    print(submission.head(3))
```

**KMeans**

```python
model_data = KMeans(n_clusters=13)
model_data.fit(data)
```

```python
make_sub(model_data.labels_)
```

**AgglomerativeClustering (очень долго обучается и показывает плохой результат)**

```python
model_AC = AgglomerativeClustering(n_clusters=13, linkage='ward')
model_AC.fit(data.toarray())
```

```python
make_sub(model_AC.labels_)
```

**DBSCAN (лучший результат, eps=1.3425, min_samples=7)**

```python
model_db = DBSCAN(eps=1.3425, min_samples=7)
model_db.fit(data)
```

```python
submission = pd.DataFrame({'ID':range(data.shape[0]), 'TARGET':model_db.labels_})
print(submission.query('TARGET == -1').TARGET.count())
submission.describe()
```

```python
make_sub(model_db.labels_)
```

**HDBSCAN (min_cluster_size=3)**

```python
model_hdb = hdbscan.HDBSCAN(min_cluster_size=3)
model_hdb.fit(data)
```

```python
submission = pd.DataFrame({'ID':range(data.shape[0]), 'TARGET':model_hdb.labels_})
print(submission.query('TARGET == -1').TARGET.count())
submission.describe()
```

```python
make_sub(model_hdb.labels_)
```
